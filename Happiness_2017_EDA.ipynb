# Happiness 2017 Dataset â€“ EDA & Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("happiness_2017.csv")
df.head()

```

## Descriptive Statistics

```python
df.info()
df.describe()
```

## Visualizations

```python

numeric_df = df.select_dtypes(include=[np.number])
numeric_df.head()

```

```python
numeric_df.hist(figsize=(12, 8))
plt.suptitle("Histogram of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=numeric_df)
plt.xticks(rotation=45)
plt.title("Box Plot of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(numeric_df)
plt.show()
```

## Observations

1. Strong correlation is observed between economy-related features and the happiness score.
2. Some numerical features contain outliers, as identified from the box plot.
3. Histograms indicate that the features follow different data distributions across countries.
